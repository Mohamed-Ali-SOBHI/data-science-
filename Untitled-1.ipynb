```python
import pandas as pd
```

```python
#lecture du fichier 
```

```python
fd = pd.read_csv("all_data.csv")
fd.head()
```

```python
#suprimer les valeurs nan
```

```python
nan_df = fd[fd.isna().any(axis=1)]
nan_df.head
fd = fd.dropna(how='all')
```

```python
#suprimer les valeurs or
```

```python
fd = fd[fd['Order Date'].str[0:2] != 'Or']
```

```python
#ajouter une colonne Month
```

```python
fd['Month'] = fd['Order Date'].str[0:2]
fd['Month'] = fd['Month'].astype('int32')
```

```python
#ajouter une colonne city
```

```python
fd['city'] = fd['Purchase Address'].apply(lambda x: x.split(',')[1])
fd.head()
```

```python
#convertir les données 
```

```python
fd['Quantity Ordered'] = pd.to_numeric(fd['Quantity Ordered'])
fd['Price Each'] = pd.to_numeric(fd['Price Each'])
```

```python
#ajouter une colonne sales
```

```python
fd['sales'] = fd['Quantity Ordered'] * fd['Price Each']
fd.head()
```

```python
#récupérer le meilleur mois de vente 
```

```python
result = fd.groupby('Month').sum()
result
```

```python
#genere un graphe
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
plt.bar(months,result['sales'])
plt.xticks(months)
plt.xlabel('mois')
plt.ylabel('vente en millions de $')
plt.show()
```

```python
#récupérer la ville avec les mellieurs ventes
```

```python
result = fd.groupby('city').sum()
result
```

```python
cities = [city for city, fd in fd.groupby('city')]
plt.bar(cities,result['sales'])
plt.xticks(cities, rotation='vertical', size=10)
plt.xlabel('city name')
plt.ylabel('vente en millions de $')
plt.show()
```

```python
fd['Order Date'] = pd.to_datetime(fd['Order Date'])
```

```python
#ajouté une colonne hour
```

```python
fd['hour'] = fd['Order Date'].dt.hour
fd['minute'] = fd['Order Date'].dt.minute
fd.head()
```

```python
result = fd.groupby('hour').count()
result
```

```python
hour = range(0,24)
plt.plot(hour,result['sales'])
plt.xticks(hour)
plt.grid()
plt.xlabel('hour')
plt.ylabel('vente en $')
plt.show()
```
